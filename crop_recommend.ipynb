```python
import pandas as pd
```

```python
df = pd.read_csv('Crop_recommendation.csv')
df.head(6)
```

```python
df.isnull().sum()    #check if there is any null value and sum tells all value at once
```

```python
df.info()
```

```python
x = df.drop('label', axis=1)
y = df['label']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 1,test_size=0.2)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)      #since we are training here so we are only using train dataset
```

```python
y_pred1=model.predict(x_test)   #here we are going to test this dataset so we will be using only test dataset
```

```python
from sklearn.metrics import accuracy_score
logistic_reg_acc=accuracy_score(y_test, y_pred1)
print("logistic accuracy is "+ str(logistic_reg_acc))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(x_train, y_train)
y_pred3= model2.predict(x_test)
```

```python
decision_acc=accuracy_score(y_test, y_pred3)
print("decision accuracy is "+ str(decision_acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3= RandomForestClassifier()
model3.fit(x_train, y_train)
y_pred4=model3.predict(x_test)
```

```python
random_acc = accuracy_score(y_test, y_pred4)
print("random forest accuracy is "+ str(random_acc))
```
